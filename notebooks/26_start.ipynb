```python
# grab the API key
with open('/content/my_key.txt', 'r') as file:
  api_key = file.read().strip()

# open API key
os.environ["OPENAI_API_KEY"] = api_key

openai.api_key = api_key
```

```python
prompt = input("What meal are you planning for? Breakfast, Lunch or Dinner? ")

response = openai.Completion.create(
    model = "text-davinci-003",
    prompt = "Act like my personal chef and give me food suggesstions based upon the following prompt" + prompt,
    temperature = 0,
    max_tokens = 100,
    stop = None
)

message = response.choices[0].text

print(message)
```
