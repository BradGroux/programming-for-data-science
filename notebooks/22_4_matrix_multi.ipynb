```r
# import motortrend car data
data(mtcars)
```

```r
# create a matrix
mtcars_matrix <- as.matrix(mtcars)
mtcars_matrix
```

#Matrix Operations

```r
# matrix transposition
mtcars_matrix_transpose <- t(mtcars_matrix)
mtcars_matrix_transpose
```

```r
# matrix multiply
result <- mtcars_matrix %*% mtcars_matrix_transpose
result
```
