```r
# vector using the c-function
my_nums <- c(1,2,3,4,5)
my_nums
```

```r
# calling characters in r (strings in python)
my_char_vector <- c("a", "b", "c", "d", "e")
my_char_vector
```

```r
# vector is a one dimensional array of multiple data types
fruits <- c("apple", "banana", "banana", "apple", "banana")

# Factors are categorical variables that take on a limited number of values
fruits_factor <- factor(fruits)
fruits_factor
```

```r
table(fruits_factor)
```

```r
#create a table using table()
col1 <- c("a", "b", "c")
col2 <- c(1, 2, 3)

table_data <- data.frame(col1, col2)
table_data
```

```r
# list - the same data types is a vector, different data types is a list
my_list <- list(1, "a", TRUE, c(2,3,4))
my_list
```

```r
# vector - the same data types is a vector, different data types is a list
my_list <- list(1,2,3,4)
my_list
```

```r
# dataframe
my_data_frame <- data.frame(x = 1:3, y = c("a", "b", "c"))
my_data_frame
```
