```r
# weight and mpg comparison
plot(mtcars$wt, mtcars$mpg)
fit <- lm(mpg ~ wt, data=mtcars)
abline(fit)
```

```r
# cyliders and mpg comparison
data(mtcars)
plot(mtcars$mpg, mtcars$cyl, xlab="MPG", ylab="Cyl", main="Relation between MPH and Cylinders")
```
