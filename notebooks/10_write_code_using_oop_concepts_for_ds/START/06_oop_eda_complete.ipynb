# Import and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Raw Dataset

```python
url = 'https://raw.githubusercontent.com/jonathan-data-analysis/colab_test_data/main/athletes.csv'

# convert to dataframe
df = pd.read_csv(url)
```

## Class Object

```python
class Summary:
    def __init__(self, data): #initializes the instance of the class
        self.df = pd.read_csv(url)
    def print_head(self, rows=None): #print first 5 rows
        return self.df.head(rows)
    def get_columns(self): #print column headers
        return list(self.df.columns)
    def get_dimensions(self): #print number of rows and columns
        print('Rows:' , len(self.df))
        print('Columns:' , len(self.df.columns))
    def get_stats(self): #print summary statistics
        return self.df.describe()
    def get_mean(self, column): #print mean of a column
        print(f"Mean {column}: {self.df[column].mean()}")
    def get_standard_dev(self, column): #print standard deviation of a column
        print(f"Standard Deviation {column}: {self.df[column].std()}")
    def get_histogram(self, column): #data visualization: histogram
        sns.set()
        self.df[column].hist(bins=10)
        plt.title(f'Histogram of {column}')
        plt.show()
```

## Call Method Using a Variable Name `data`

```python
# assign class with raw_data to varible name data
data = Summary(url)
```
